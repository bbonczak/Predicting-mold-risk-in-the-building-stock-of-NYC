## Mold Violation Dataset Preprocessing
This notebook is for filtering and preprocessing the Mold Violations DAtaset obtained through: https://data.cityofnewyork.us/Housing-Development/MOLD-VIOLATIONS-RECENT/wguz-g589/data

THe idea is to use this dataset to identify the location, building ID of buildings that have or had mold in their past down to 2010. 
COlumns exported: 'BBL','InspectionDate','Latitude','Longitude','Postcode','CurrentStatus','CurrentStatusDate'

```python
#Import stuff
import pandas as pd
import numpy as np
```

```python
#import Dataset as Dataframe
Mold_violations=pd.read_csv('..\Data\MOLD_VIOLATIONS-_RECENT.csv')
```

```python
Mold_violations.head()
```

```python
# Mold_violations.isna().any()
```

```python
#selected potentially important columns. filter dates for everything newer 2010. 
Mold_violations=Mold_violations[['BBL','InspectionDate','Latitude','Longitude','Postcode','CurrentStatus','CurrentStatusDate']].loc[Mold_violations.InspectionDate >= '01/01/2010']
#Switch InspectionDate to pd datetime 
Mold_violations.InspectionDate = pd.to_datetime(Mold_violations.InspectionDate)
Mold_violations=Mold_violations.sort_values(by='InspectionDate')
print('length of df before dropNA: ',Mold_violations.size)
#drop Nan values
Mold_violations=Mold_violations.dropna()
print('length of df after dropNA: ',Mold_violations.size)
#switch bbl,postcode to int
Mold_violations[['BBL','Postcode']]= Mold_violations[['BBL','Postcode']].astype('int')
```

```python
#Check for remaining null values
Mold_violations[pd.isnull(Mold_violations).any(axis=1)]
```

```python
#write Csv
Mold_violations.to_csv('../Output/MoldViolationsCleaned.csv')
```
